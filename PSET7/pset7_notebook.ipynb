```python
import numpy as np
from matplotlib import pyplot as plt
from numpy.random import standard_cauchy

```

```python
x = np.linspace(0,10,1000)
```

```python
def gaussian(x,a,mu,sig):
    return np.exp(-0.5*(x-mu)**2/sig**2)
```

```python
def lorentzian(x,a):
    return a/(1+x**2)
```

```python
def power(x,a,b):
    
    return a*x**-0.05
```

```python
plt.plot(x,np.exp(-x))
plt.plot(x,gaussian(x,1,0,3.5))
```

```python
plt.plot(x,np.exp(-x))
plt.plot(x,lorentzian(x,1))
```

```python
plt.plot(x[1:],np.exp(-x[1:]))
plt.plot(x[1:],power(x[1:],1,1))
```

```python
N = 1000000
z = abs(standard_cauchy(N))
z = z[z <= 10]
N = len(z)
qz = lorentzian(z,1)
pz = np.exp(-z)
u = np.random.rand(N)*qz
accept = u<=pz
z = z[accept]

print("Fraction of accepted draws: ",np.mean(accept))
a,b=np.histogram(z, bins=x)
plt.bar(b[0:-1],a/np.max(a), width=x[1]-x[0])
plt.plot(x, lorentzian(x,1), color = "purple")
plt.plot(x, np.exp(-x), color = "orange")
```

```python
N = 1000000
u = np.random.rand(N)
v = np.random.rand(N)*2/np.exp(1)

z = v/u

accept  = u <= np.sqrt(np.exp(-z))
z = z[accept]

print("Fraction of accepted draws: ",np.mean(accept))
a,b=np.histogram(z, bins=x)
plt.bar(b[0:-1],a/np.max(a), width=x[1]-x[0])
plt.plot(x, np.exp(-x), color = "orange")
```
