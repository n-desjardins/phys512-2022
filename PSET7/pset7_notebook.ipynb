Q2

```python
x = np.linspace(0,10,1000)
```

```python
def gaussian(x,a,mu,sig):
    return np.exp(-0.5*(x-mu)**2/sig**2)
```

```python
def lorentzian(x):
    return 1/(1+1*x**2)
```

```python
def power(x):
    
    return x**-0.05
```

```python
plt.plot(x,np.exp(-x), label="exponential")
plt.plot(x,gaussian(x,1,0,3.5), label="gaussian")
plt.legend()
plt.savefig("exp.png")
```

```python
plt.plot(x,np.exp(-x), label="exponential")
plt.plot(x,lorentzian(x), label="lorentzian")
plt.legend()
plt.savefig("lorentz.png")
```

```python
plt.plot(x,np.exp(-x), label="exponential")
plt.plot(x,power(x), label="power law")
plt.legend()
plt.savefig("power.png")
```

```python
N = 1000000
z = abs(standard_cauchy(N))
qz = lorentzian(z)
pz = np.exp(-z)
u = np.random.rand(N)*qz
accept = u<=pz
z = z[accept]

print("Fraction of accepted draws: ",np.mean(accept))
a,b=np.histogram(z, bins=x)
plt.bar(b[0:-1],a/np.max(a), width=x[1]-x[0])
plt.plot(x, lorentzian(x), color = "purple", label="Lorentzian")
plt.plot(x, np.exp(-x), color = "orange", label="exponential")
plt.title("Q2: using the rejection method, efficiency = 0.636388")
plt.xlabel("x")
plt.ylabel("PDF(x)")
plt.legend()
plt.savefig("Q2.png")
```

```python
N = 1000000
u = np.random.rand(N)
v = np.random.rand(N)*2/np.exp(1)

z = v/u

accept  = u <= np.sqrt(np.exp(-z))
z = z[accept]

print("Fraction of accepted draws: ",np.mean(accept))
a,b=np.histogram(z, bins=x)
plt.bar(b[0:-1],a/np.max(a), width=x[1]-x[0])
plt.plot(x, np.exp(-x), color = "orange", label="exponential")
plt.title("Q3: using the ratio-ofuniforms, efficiency = 0.679961")
plt.xlabel("x")
plt.ylabel("PDF(x)")
plt.legend()
plt.savefig("Q3.png")
```
